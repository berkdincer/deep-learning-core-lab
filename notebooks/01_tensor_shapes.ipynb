```
# Create a 2D tensor

x_2d = torch.rand((4,4))

print(x_2d)
print("\n")

# Access the element at row 1, column 2
print(x_2d[0,1])
print("\n")

# Access the entire second column (index 1)
print(x_2d[:,1])
print("\n")

# Modify an element
x_2d[1,1] = 2
print(x_2d)
print("\n")

# Select the first two rows and columns 1 and 2
print(x_2d[0:2, 1:3])
print("\n")

# Select all rows, but only the last two columns
print(x_2d[:, -2:])
print("\n")
```

## Shape Manipulation

Let's continue with shape manipulation, specifically view vs reshape. The main deal is: view uses the exact same data directly, while reshape creates a brand new copy if it has to.

This all comes down to the logic of 'contiguity'. If the data in the tensor is lined up perfectly back-to-back in the memory (RAM), we can easily grab it using view. But if there are weird, random gaps in the memory layout, we can't do that.

For example, if we try to grab the 3rd and 5th columns, the data isn't sitting together—it's jumping around in the memory. Since it's not contiguous anymore, view fails, and we are forced to use reshape so it can create a fresh, straight copy of that data under the hood

```
# view, reshape


v_tensor = torch.arange(12)
print(v_tensor)

a = v_tensor.view((3,4))
print(a)

# to make flatten -1 must be written such that:

flatten_shape = a.view(-1)
print(flatten_shape)
```

```
# Inplace method: Direct change the memory element. Do not create any copy.

t2 = torch.tensor([1,2,3])
t2.add_(5) # _ -> generally, use for in-place methods.
print(t2)
```
